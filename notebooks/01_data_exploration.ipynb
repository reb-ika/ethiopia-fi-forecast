```python
# Fix ID collision: events get the next IDs after all the observations
new_evt_ids = next_ids(pd.DataFrame({"record_id": data_df["record_id"].tolist() + new_obs_ids}),
                        "REC", len(new_events))

for rec, rid in zip(new_observations, new_obs_ids):
    rec["record_id"] = rid
for rec, rid in zip(new_events, new_evt_ids):
    rec["record_id"] = rid

print("Event ID:", new_evt_ids)  # will need this for the impact_link parent_id

# Merge into main dataset
enriched_data_df = pd.concat(
    [data_df, pd.DataFrame(new_observations), pd.DataFrame(new_events)],
    ignore_index=True
)

print("Original data_df shape:", data_df.shape)
print("Enriched data_df shape:", enriched_data_df.shape)
```

```python
def make_impact_record(**kwargs):
    row = {col: np.nan for col in impact_df.columns}
    row.update(kwargs)
    return row

new_impact_ids = next_ids(pd.DataFrame({"record_id": impact_df["record_id"].tolist()}), "IMP", 3)

new_impact_links = [
    # Closes the gap: EVT_0006 (P2P/ATM crossover milestone) had no impact_link before
    make_impact_record(record_id=new_impact_ids[0], parent_id="EVT_0006", record_type="impact_link",
                        pillar="USAGE", related_indicator="USG_CROSSOVER", relationship_type="milestone_confirmation",
                        impact_direction="increase", impact_magnitude="high", lag_months=0,
                        evidence_basis="empirical", collected_by="Rebe", collection_date=today,
                        source_name="EthSwitch Annual Report / Capital Newspaper",
                        source_url="https://capitalethiopia.com/2025/11/02/ethswitch-reports-historic-growth-as-p2p-payments-surpass-atm-withdrawals/",
                        confidence="high",
                        notes="P2P interoperable transaction volume exceeded ATM withdrawal volume for first time, Oct 2024; P2P value also surpassed ATM value in FY24/25"),
    # New EthioPay launch event -> likely effects (theoretical, not yet observed - it's a Dec 2025 launch)
    make_impact_record(record_id=new_impact_ids[1], parent_id="REC_0044", record_type="impact_link",
                        pillar="USAGE", related_indicator="USG_P2P_COUNT", relationship_type="event_impact",
                        impact_direction="increase", impact_magnitude="medium", lag_months=6,
                        evidence_basis="theoretical", collected_by="Rebe", collection_date=today,
                        source_name="Capital Newspaper",
                        source_url="https://capitalethiopia.com/2025/12/21/ethswitch-launches-ethiopay-to-unify-national-digital-paymentsby-our-staff-reporter/",
                        confidence="low",
                        notes="EthioPay brand unification launched Dec 2025; too recent for empirical evidence, estimate is theoretical based on stated goal of unifying fragmented digital payment services"),
    make_impact_record(record_id=new_impact_ids[2], parent_id="REC_0044", record_type="impact_link",
                        pillar="ACCESS", related_indicator="ACC_MM_ACCOUNT", relationship_type="event_impact",
                        impact_direction="increase", impact_magnitude="low", lag_months=12,
                        evidence_basis="theoretical", collected_by="Rebe", collection_date=today,
                        source_name="Capital Newspaper",
                        source_url="https://capitalethiopia.com/2025/12/21/ethswitch-launches-ethiopay-to-unify-national-digital-paymentsby-our-staff-reporter/",
                        confidence="low",
                        notes="Unified branding may reduce friction/confusion across mobile money providers, modestly aiding account adoption"),
]

enriched_impact_df = pd.concat([impact_df, pd.DataFrame(new_impact_links)], ignore_index=True)

print("Original impact_df shape:", impact_df.shape)
print("Enriched impact_df shape:", enriched_impact_df.shape)

# Save to data/processed/
processed_dir = Path("../data/processed")
processed_dir.mkdir(exist_ok=True)

with pd.ExcelWriter(processed_dir / "ethiopia_fi_unified_data_enriched.xlsx") as writer:
    enriched_data_df.to_excel(writer, sheet_name="ethiopia_fi_unified_data", index=False)
    enriched_impact_df.to_excel(writer, sheet_name="Impact_sheet", index=False)

print("Saved to:", processed_dir / "ethiopia_fi_unified_data_enriched.xlsx")
```
